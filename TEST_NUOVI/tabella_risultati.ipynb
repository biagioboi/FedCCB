```python
import json
from pathlib import Path
import pandas as pd

distribuzioni = ["0.1", "0.5", "0.9"]
attacchi = ["basso", "medio", "alto"]

#  diz[dist][att][metodo] 
diz = {}

rows = []

for dist in distribuzioni:
    base_path = Path(f"./distribuzione {dist}") / "results"

    for att in attacchi:
        results_file = base_path / att / "results.txt"

        if not results_file.exists():
            print(f"File non trovato: {results_file}")
            continue

        with results_file.open("r", encoding="utf-8") as f:
            dizionario = json.load(f)

        diz.setdefault(dist, {})
        diz[dist].setdefault(att, {})

        for metodo, rounds_dict in dizionario.items():
            print(dist,metodo,att)
            last_key = max(rounds_dict.keys(), key=lambda k: int(k))
            statistiche = rounds_dict[last_key]

            diz[dist][att][metodo] = statistiche

            row = {
                "distribuzione": dist,
                "attacco": att,
                "metodo": metodo,
                "round": int(last_key),
            }
            row.update(statistiche)
            rows.append(row)

df = pd.DataFrame(rows)

if not df.empty:
    df = df.sort_values(["distribuzione", "attacco", "metodo"]).reset_index(drop=True)

df.to_csv("./tabella.csv")
```
